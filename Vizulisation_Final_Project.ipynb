```python
df["date_added"] = df["date_added"].str.strip()
df["date_added"] = pd.to_datetime(df["date_added"], errors='coerce', format="%B %d, %Y")
```

```python
disney_movies = df[df.type == 'Movie']
disney_shows = df[df.type == 'TV Show']
df['year'] = df['date_added'].dt.year.astype(int)
df['month'] = df['date_added'].dt.month.astype(int)
df['release_year'] = df['release_year'].astype(int)
df
```

```python
# Function to extract season count
def extract_season_count(x):
    if isinstance(x, str) and "Season" in x:
        return x.split(" ")[0]
    else:
        return ""

# Function to extract duration
def extract_duration(x):
    if isinstance(x, str) and "Season" not in x:
        return x.split(" ")[0]
    else:
        return ""

# Apply the functions to create new columns
df['season_count'] = df['duration'].apply(extract_season_count)
df['duration'] = df['duration'].apply(extract_duration)

df
```

```python
from collections import Counter
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

# Combine titles and descriptions into a single text column
df['combined_text'] = df['title'].fillna('') + ' ' + df['description'].fillna('')
stop_words = set(stopwords.words('english'))

# Function to clean and tokenize text
def clean_and_tokenize(text):
    # Convert text to lowercase
    text = text.lower()
    # Remove punctuation and special characters
    text = re.sub(r'[^\w\s]', '', text)
    # Tokenize the text
    tokens = text.split()
    tokens = [token for token in tokens if token not in stop_words]
    return tokens

# Apply clean_and_tokenize function to combined_text column
df['tokens'] = df['combined_text'].apply(clean_and_tokenize)

# Flatten the list of tokens
all_tokens = [token for sublist in df['tokens'] for token in sublist]

# Count the occurrence of each token
word_counts = Counter(all_tokens)

# Sort the words by their frequency
most_common_words = word_counts.most_common()
```

```python
# For movies
df_movies = df[df['type'] == 'Movie']

# Combine titles and descriptions into a single text column for movies
df_movies['combined_text'] = df_movies['title'].fillna('') + ' ' + df_movies['description'].fillna('')

# Apply clean_and_tokenize function to combined_text column for movies
df_movies['tokens'] = df_movies['combined_text'].apply(clean_and_tokenize)

# Flatten the list of tokens for movies
all_tokens_movies = [token for sublist in df_movies['tokens'] for token in sublist]

# Count the occurrence of each token for movies
word_counts_movies = Counter(all_tokens_movies)

# Sort the words by their frequency for movies
most_common_words_movies = word_counts_movies.most_common()

# For shows
df_shows = df[df['type'] == 'TV Show']

# Combine titles and descriptions into a single text column for shows
df_shows['combined_text'] = df_shows['title'].fillna('') + ' ' + df_shows['description'].fillna('')

# Apply clean_and_tokenize function to combined_text column for shows
df_shows['tokens'] = df_shows['combined_text'].apply(clean_and_tokenize)

# Flatten the list of tokens for shows
all_tokens_shows = [token for sublist in df_shows['tokens'] for token in sublist]

# Count the occurrence of each token for shows
word_counts_shows = Counter(all_tokens_shows)

# Sort the words by their frequency for shows
most_common_words_shows = word_counts_shows.most_common()
```
